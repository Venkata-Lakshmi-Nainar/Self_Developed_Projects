```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
print(boston)
```

```python
print(boston['feature_names'])
```

```python
print(boston['DESCR'])
```

```python
boston_df = pd.DataFrame(boston['data'])
```

```python
boston_df
```

```python
boston_df.columns = pd.DataFrame(boston['feature_names'])[0].copy()
boston_df['target'] = pd.DataFrame(boston['target'])[0].copy()
```

```python
boston_df.head()
```

# Split Data X and Y

```python
from sklearn.model_selection import train_test_split
```

```python
X = boston_df.copy()
X.drop(columns=['target'], inplace=True)
X
```

```python
Y = pd.DataFrame(boston_df['target'].copy())
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X,Y)
lin_reg_result = pd.DataFrame(lin_reg.predict(X))
lin_reg_result.rename(columns={lin_reg_result.columns[0]:'Pred_y'}, inplace=True)
lin_reg_result
```

```python
lin_reg_result['Actual_y'] = Y['target'].copy()
lin_reg_result = lin_reg_result.round(decimals=1)
lin_reg_result
```

#### Accuracy Metrics

```python
import sklearn.metrics as metrics
```

```python
mae = metrics.mean_absolute_error(lin_reg_result['Pred_y'], lin_reg_result['Actual_y'])
mse = metrics.mean_squared_error(lin_reg_result['Pred_y'], lin_reg_result['Actual_y'])
r2 = metrics.r2_score(lin_reg_result['Pred_y'], lin_reg_result['Actual_y'])
print("Mean Absolute Error: ", mae)
print("Mean Squared Error: ", mse)
print("R-Squared:", r2)
```

```python
for col in X.columns:
    print('Feature: ',col)
    plt.scatter(X[col], lin_reg_result['Pred_y'], color="blue", label="original")
    plt.scatter(X[col], lin_reg_result['Actual_y'], s=15, color="red", label="predicted")
    plt.legend()
    plt.show()
```

## Random Forest

```python
print("X-Train: ", X_train.shape)
print("Y-Train: ", Y_train.shape)
print("X-Test: ", X_test.shape)
print("Y-Test: ", Y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
classifier_rf = RandomForestRegressor(random_state=42, n_jobs=-1, max_depth=5, n_estimators=100, oob_score=True)
classifier_rf.fit(X_train, np.array(Y_train).ravel())
```

```python
ran_for_result = pd.DataFrame(classifier_rf.predict(X_test)).copy()
ran_for_result.rename(columns={ran_for_result.columns[0]: 'Pred_y'}, inplace=True)
ran_for_result['Actual_y'] = (Y_test.reset_index())['target'].copy()
ran_for_result
```

```python
for col in X.columns:
    print('Feature: ',col)
    plt.scatter(X_test[col], ran_for_result['Pred_y'], color="blue", label="original")
    plt.scatter(X_test[col], ran_for_result['Actual_y'], s=15, color="red", label="predicted")
    plt.legend()
    plt.show()
```

#### Accuracy Metrics

```python
mae = metrics.mean_absolute_error(ran_for_result['Pred_y'], ran_for_result['Actual_y'])
mse = metrics.mean_squared_error(ran_for_result['Pred_y'], ran_for_result['Actual_y'])
r2 = metrics.r2_score(ran_for_result['Pred_y'], ran_for_result['Actual_y'])
print("Mean Absolute Error: ", mae)
print("Mean Squared Error: ", mse)
print("R-Squared:", r2)
```
